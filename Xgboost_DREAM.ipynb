```python
anoSC1 = pd.read_csv('/Users/User/Downloads/anoSC1_v11_nokey.csv.xls') 
anoSC1 = anoSC1.set_index('SampleID')
geneExp = pd.read_csv('/Users/User/Downloads/csv_genes.csv')
geneExp = geneExp.set_index('Unnamed: 0')
geneExp = geneExp.T
geneExp = pd.concat([geneExp, anoSC1], axis=1, sort=False)
geneExp = geneExp.drop(['Batch','Set','Platform'],axis=1)
#geneExp
geneExpdata= geneExp.loc[geneExp['Train'] == 1]
geneExpdata
X = geneExpdata.iloc[:,0:32830]
y = geneExpdata.iloc[:,-2]



```
